```python
total = pd.read_csv('total.csv', index_col = 0)
```

```python
#make lists of top 20 long/short tickers
top20long = total[total.Strategy == 'Long'].groupby('Ticker').Ticker.count().sort_values(ascending = False).head(20)
top20short = total[total.Strategy == 'Short'].groupby('Ticker').Ticker.count().sort_values(ascending = False).head(20)
```

```python
# returns for both long and short strategy for boxplot, excluding outliers for plotting purposes
box_data_long = total[total.Strategy == 'Long'][['Week_Return', 'Month_Return', 'Quarter_Return', 'Semi_Return', 'Annual_Return']]
box_data_long = box_data_long.loc[box_data_long.apply(lambda x: np.abs(x - x.mean()) / x.std() < 2.5).all(axis=1)]

box_data_short = total[total.Strategy == 'Short'][['Week_Return', 'Month_Return', 'Quarter_Return', 'Semi_Return', 'Annual_Return']]
box_data_short = box_data_short.loc[box_data_short.apply(lambda x: np.abs(x - x.mean()) / x.std() < 2.5).all(axis=1)]
```

```python
#plotly boxplot - stock returns by different strategies and holding periods
trace0 = go.Box(
    y = box_data_long.Week_Return,
    name = 'Long - 7days',
    boxpoints = 'all',
    marker = dict(
        color='rgb(40, 180, 208)',
    ),
    boxmean='sd'
)
trace1 = go.Box(
    y = box_data_short.Week_Return,
    name = 'Short - 7days',
    boxpoints = 'all',
    marker = dict(
        color='rgb(400, 40, 40)',
    ),
    boxmean='sd'
)
trace2 = go.Box(
    y = box_data_long.Month_Return,
    name = 'Long - 30days',
    boxpoints = 'all',
    marker = dict(
        color='rgb(20, 150, 208)',
    ),
    boxmean='sd'
)
trace3 = go.Box(
    y = box_data_short.Month_Return,
    name = 'Short - 30days',
    boxpoints = 'all',
    marker = dict(
        color='rgb(250, 40, 100)',
    ),
    boxmean='sd'
)
trace4 = go.Box(
    y = box_data_long.Quarter_Return,
    name = 'Long - 90days',
    boxpoints = 'all',
    marker = dict(
        color='rgb(20, 100, 180)',
    ),
    boxmean='sd'
)
trace5 = go.Box(
    y = box_data_short.Quarter_Return,
    name = 'Short - 90days',
    boxpoints = 'all',
    marker = dict(
        color='rgb(200, 40, 120)',
    ),
    boxmean='sd'
)
trace6 = go.Box(
    y = box_data_long.Annual_Return,
    name = 'Long - 360days',
    boxpoints = 'all',
    marker = dict(
        color='rgb(8, 81, 156)',
    ),
    boxmean='sd'
)
trace7 = go.Box(
    y = box_data_short.Annual_Return,
    name = 'Short - 360days',
    boxpoints = 'all',
    marker = dict(
        color='rgb(150, 40, 150)',
    ),
    boxmean='sd'
)
data = [trace0, trace1, trace2, trace3, trace4, trace5, trace6, trace7]
layout = go.Layout(
    title='Stock Returns by Strategy/Holding Period',
    xaxis=dict(title = 'Strategy/Holding Period',
              autorange=True,
        showgrid=True,
        zeroline=True,
        zerolinewidth=2),
    yaxis=dict(
        title = 'Returns',
        autorange=True,
        showgrid=True,
        zeroline=True,
        zerolinewidth=2))

fig = go.Figure(data=data, layout=layout)

py.plot(fig, show_link=False, image_width=1000, image_height=800, filename = 'Long_Short_Boxplot.html')
```
